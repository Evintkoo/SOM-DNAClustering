```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class quadatic_function():
    def __init__(self, a, b, c) -> None:
        self.a = a
        self.b = b 
        self.c = c
    
    def predict(self, X):
        return self.a*X*X + self.b*X + self.c
    
    def peak_point(self):
        return (-self.b/(self.a*2), (self.b*self.b - 4*self.a*self.c)/(4*self.a))
```

```python
x = [-1, 0, 1, 2, 3, 4]
fn = quadatic_function(1,1,2)
y = [fn.predict(i) for i in x]
print(fn.peak_point())
plt.scatter(x,y)
```

```python
coefficients = np.polyfit(x, y, 2)
model = quadatic_function(coefficients[0], coefficients[1], coefficients[2])
print(model.a, model.b, model.c)
print(model.peak_point())
```
